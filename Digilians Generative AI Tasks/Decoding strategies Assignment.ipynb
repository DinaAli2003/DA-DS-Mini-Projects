```python
# Import Libraries and Configure API Key
from google import genai
from google.genai.types import GenerateContentConfig

# Add Gemini API Key
API_KEY = "AIzaSyAyCz1lecaH3jlugXl6d5gwhHjcPfa09r8"

# Create client
client = genai.Client(api_key=API_KEY)
```

```python
# Define the Prompt
prompt = """
Suggest 3 innovative names for a student study mobile app,
formatted as a numbered list,
and explain each in one line.
"""
```

```python
# Create a Helper Function
def run_profile(profile_name, config):

    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=prompt,
        config=config
    )

    print("\n" + "=" * 60)
    print(profile_name)
    print("=" * 60)

    print(response.text)
```

```python
# Profile 1: The Strict Employee
strict_config = GenerateContentConfig(
    temperature=0
)

run_profile("Profile 1 - The Strict Employee", strict_config)
```

```python
# Profile 2: The Creative Marketer
creative_config = GenerateContentConfig(
    temperature=1.5,
    top_p=0.9
)

run_profile("Profile 2 - The Creative Marketer", creative_config)
```

```python
# Profile 3: The Budget Manager
budget_config = GenerateContentConfig(
    max_output_tokens=25
)

run_profile("Profile 3 - The Budget Manager", budget_config)
```

```python
# Profile 4: The Interrupter
interrupter_config = GenerateContentConfig(
    temperature=0,
    stop_sequences=["3."]
)

run_profile("Profile 4 - The Interrupter", interrupter_config)
```

```python
# Profile 5: The Focused Expert
focused_config = GenerateContentConfig(
    temperature=1.2,
    top_k=5
)

run_profile("Profile 5 - The Focused Expert", focused_config)
```

**Profile 1**

I used temperature=0 to remove randomness and make the output deterministic.

**Profile 2**

I increased creativity using high temperature and controlled coherence using top_p.

**Profile 3**

I used max_output_tokens to reduce response length and save cost.

**Profile 4**

I used stop_sequences to force generation to stop when the model reached “3.”.

**Profile 5**

I allowed creativity but constrained token selection using top_k=5.
