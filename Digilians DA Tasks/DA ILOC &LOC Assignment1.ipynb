Read csv file and apply these steps:

•Select rows 1 to 3 and columns Customer and Amount using iloc.

•Select only the Amount column using iloc.

•Get the value of City in the 3rd row using iloc.

•Select Customer and Amount for rows where Amount > 300.

```python


import pandas as pd

# Create dictionary
data = {
    "OrderID": [101, 102, 103, 104, 105],
    "Customer": ["Ali", "Sara", "Omar", "Mona", "Youssef"],
    "City": ["Cairo", "Giza", "Alex", "Cairo", "Giza"],
    "Amount": [250, 400, 150, 300, 500],
    "OrderDate": [
        "2024-01-10",
        "2024-01-12",
        "2024-02-05",
        "2024-02-20",
        "2024-03-01"
    ]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Convert OrderDate to datetime
df["OrderDate"] = pd.to_datetime(df["OrderDate"])
```

```python
df.to_csv("orders.csv", index=False)

df = pd.read_csv("orders.csv")
df["OrderDate"] = pd.to_datetime(df["OrderDate"])
```

```python
# Select first 3 rows
df.iloc[:3]
```

```python
#Select last row
df.iloc[-1]
```

```python
#Rows 1–3 & columns Customer and Amount
df.iloc[1:4, [1, 3]]
```

```python
#Select only Amount column
df.iloc[:, 3]
```

```python
#City value in 3rd row
df.iloc[2]["City"]
```

```python
#Rows where City = Cairo
df[df["City"] == "Cairo"]
```

```python
#Customer & Amount where Amount > 300
df.loc[df["Amount"] > 300, ["Customer", "Amount"]]
```

```python
#Rows with OrderID 102 and 105
df[df["OrderID"].isin([102, 105])]
```

```python
#Update Amount to 600 where Customer = Sara
df.loc[df["Customer"] == "Sara", "Amount"] = 600
```

```python
#City is NOT Giza
df[df["City"] != "Giza"]
```

```python
#Amount between 200 and 400
df[df["Amount"].between(200, 400)]
```

```python
#City Cairo OR Giza
df[df["City"].isin(["Cairo", "Giza"])]
```

```python
#Customer name starts with "M"
df[df["Customer"].str.startswith("M")]
```

```python
#City Cairo AND Amount > 250
df[(df["City"] == "Cairo") & (df["Amount"] > 250)]
```

```python
#Sort by Amount (descending) & top 3
df.sort_values(by="Amount", ascending=False).head(3)
```

```python
#Orders made in January 2024
df[(df["OrderDate"].dt.month == 1) & (df["OrderDate"].dt.year == 2024)]
```

```python
#Orders after 2024-02-01
df[df["OrderDate"] > "2024-02-01"]
```

```python
#Add Month column
df["Month"] = df["OrderDate"].dt.month
```

```python
#Orders made on Monday
df[df["OrderDate"].dt.day_name() == "Monday"]
```

```python
#Count orders per month
df.groupby(df["OrderDate"].dt.month).size()
```

```python
#First order of each month
df.sort_values("OrderDate").groupby(df["OrderDate"].dt.month).first()
```

```python
#Orders where day > 15
df[df["OrderDate"].dt.day > 15]
```

```python
#Total Amount per City per month
df.groupby([df["OrderDate"].dt.month, "City"])["Amount"].sum()
```

```python
#Orders on last day of month
df[df["OrderDate"].dt.is_month_end]
```

```python
#Orders in last 7 days of available data
last_date = df["OrderDate"].max()
df[df["OrderDate"] >= last_date - pd.Timedelta(days=7)]
```
