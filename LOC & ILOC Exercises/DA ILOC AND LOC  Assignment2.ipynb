
Read Diabetes Dataset and apply iloc and loc



```python
import pandas as pd
import numpy as np

# Read dataset
df = pd.read_csv("Diabetes Dataset.csv")

# Preview data
df.head()
```

```python
df.shape          # rows & columns
```

```python
df.columns        # column names
```

```python

df.info()         # data types & nulls
```

```python
df.describe()     # statistical summary
```

```python
#iloc & loc Operations (Index-based)
# Select first 5 rows
df.iloc[:5]
```

```python
#Select last row
df.iloc[-1]
```

```python
#Select rows 2–6 and columns 1–4
df.iloc[2:7, 1:5]
```

```python
#Select only the first column
df.iloc[:, 0]
```

```python
#Get value of Glucose in 3rd row
df.iloc[2]["Glucose"]
```

```python
#Select all rows where Outcome = 1 (Diabetic)
df.loc[df["Outcome"] == 1]
```

```python
#Select Glucose & BMI for diabetic patients
df.loc[df["Outcome"] == 1, ["Glucose", "BMI"]]
```

```python
#Select rows where Age > 40
df.loc[df["Age"] > 40]
```

```python
#Select rows where BMI > 30 AND Outcome = 1
df.loc[(df["BMI"] > 30) & (df["Outcome"] == 1)]
```

```python
#Glucose level between 100 and 140
df[df["Glucose"].between(100, 140)]
```

```python
#Patients with zero insulin (data quality check)
df[df["Insulin"] == 0]
```

```python
#Patients with BloodPressure < 80 OR BMI > 35
df[(df["BloodPressure"] < 80) | (df["BMI"] > 35)]
```

```python
#Replace zero BMI values with mean BMI
mean_bmi = df["BMI"].mean()
df.loc[df["BMI"] == 0, "BMI"] = mean_bmi
```

```python
#Sort by Glucose (descending)
df.sort_values(by="Glucose", ascending=False)
```

```python
#Top 5 patients with highest BMI
df.sort_values(by="BMI", ascending=False).head(5)
```

```python
#Count diabetic vs non-diabetic
df["Outcome"].value_counts()
```

```python
#Average Glucose per Outcome
df.groupby("Outcome")["Glucose"].mean()
```

```python
#Mean BMI and Age per Outcome
df.groupby("Outcome")[["BMI", "Age"]].mean()
```

```python
# Create Age Group column
df["AgeGroup"] = pd.cut(
    df["Age"],
    bins=[20, 30, 40, 50, 60, 100],
    labels=["20-30", "30-40", "40-50", "50-60", "60+"]
)
```

```python
#Count patients per Age Group
df["AgeGroup"].value_counts()
```

```python
#Check missing values
df.isnull().sum()
```

```python
#Count zero values per column
(df == 0).sum()
```

```python
# Diabetic patients with high glucose & BMI
df[(df["Outcome"] == 1) & (df["Glucose"] > 140) & (df["BMI"] > 30)]
```

```python
# Oldest diabetic patient
df[df["Outcome"] == 1].sort_values("Age", ascending=False).head(1)
```
